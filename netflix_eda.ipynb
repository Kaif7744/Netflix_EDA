```python
# ðŸ“¦ Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import warnings
warnings.filterwarnings('ignore')
```

```python
# ðŸ“‚ Load Dataset
df = pd.read_csv('netflix_titles.csv')
```

# Removes rows with missing title or type.

```python
# ðŸ§¹ Data Cleaning

# Drop rows with missing critical fields
df.dropna(subset=['title', 'type'], inplace=True)
```

```python
df.drop_duplicates(subset='show_id', inplace=True)  # drops based only on show_id
```

```python
df.shape
```

```python
df.describe()
```

```python
# Check for missing values in each column
print(df.isnull().sum())
```

# âœ… Fills missing values with placeholders.

```python
# Fill missing values
df['director'].fillna('Unknown', inplace=True)
df['cast'].fillna('Not Specified', inplace=True)
df['country'].fillna('Unknown', inplace=True)
df['rating'].fillna('Unknown', inplace=True)
df['date_added'].fillna('Not Available', inplace=True)
```

# Converts date strings into real datetime values, then extracts the year and month when content was added to Netflix.

```python
# Convert date_added to datetime
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
```

```python
# Extract year and month
df['year_added'] = df['date_added'].dt.year
df['month_added'] = df['date_added'].dt.month
```

# Removes duplicate records.

```python
# Drop duplicates
df.drop_duplicates(inplace=True)
```

# Basic Visualizations

```python
# ðŸ“Š EDA (Exploratory Data Analysis)

# Count of Movies vs TV Shows
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='type', palette='Set2')
plt.title('Netflix Content Type Distribution')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()
```

```python
# Content Added by Year
plt.figure(figsize=(10,5))
sns.histplot(df['year_added'].dropna(), bins=15, kde=True, color='orange')
plt.title('Netflix Content Added Over Years')
plt.xlabel('Year Added')
plt.ylabel('Number of Titles')
plt.show()
```

```python
# Top 10 Countries
top_countries = df[df['country'] != 'Unknown']['country'].value_counts().head(10)
plt.figure(figsize=(10,5))
sns.barplot(x=top_countries.values, y=top_countries.index, palette='Set3')
plt.title('Top 10 Countries on Netflix')
plt.xlabel('Number of Titles')
plt.ylabel('Country')
plt.show()
```

```python
# Rating Distribution
plt.figure(figsize=(10,5))
sns.countplot(data=df, x='rating', order=df['rating'].value_counts().index, palette='cool')
plt.xticks(rotation=45)
plt.title('Rating Distribution')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()
```

```python
# Movie Duration Distribution
movie_df = df[df['type'] == 'Movie'].copy()
movie_df['duration_int'] = movie_df['duration'].str.extract('(\d+)').astype(float)
plt.figure(figsize=(10,5))
sns.histplot(movie_df['duration_int'].dropna(), bins=30, kde=True, color='crimson')
plt.title('Movie Duration Distribution (in minutes)')
plt.xlabel('Duration (minutes)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Top 10 Directors
top_directors = df[df['director'] != 'Unknown']['director'].value_counts().head(10)
plt.figure(figsize=(10,5))
sns.barplot(x=top_directors.values, y=top_directors.index, palette='magma')
plt.title('Top 10 Directors on Netflix')
plt.xlabel('Number of Titles')
plt.ylabel('Director')
plt.show()
```

# Average Movie Duration by Genre

```python
movie_df = df[df['type'] == 'Movie'].copy()
movie_df['duration_int'] = movie_df['duration'].str.extract('(\d+)').astype(float)
movie_df['main_genre'] = movie_df['listed_in'].str.split(',').str[0]
```

```python
# Group by genre and calculate average duration
avg_duration = movie_df.groupby('main_genre')['duration_int'].mean().sort_values(ascending=False).head(10)
```

```python
# Plot
plt.figure(figsize=(10,5))
sns.barplot(x=avg_duration.values, y=avg_duration.index, palette='flare')
plt.title('Average Movie Duration by Genre')
plt.xlabel('Average Duration (min)')
plt.ylabel('Genre')
plt.show()
```

# Most Popular Genre by Country
Find out which genre is most popular in each of the top 5 countries.


```python
top_5_countries = df['country'].value_counts().head(5).index
df['main_genre'] = df['listed_in'].str.split(',').str[0]
```

```python
# Filter only top 5 countries and find mode genre
popular_genre_by_country = df[df['country'].isin(top_5_countries)].groupby('country')['main_genre'].agg(lambda x: x.mode().iloc[0])
print("Most popular genre in top 5 countries:\n", popular_genre_by_country)
```

# Correlation Between Content Release Year and Add Year
Analyze if Netflix prefers adding recent vs. older content.

```python
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df['year_added'] = df['date_added'].dt.year
df_corr = df.dropna(subset=['release_year', 'year_added'])
correlation = df_corr['release_year'].corr(df_corr['year_added'])
print("Correlation between release year and year added to Netflix:", round(correlation, 2))
```

```python
plt.figure(figsize=(8,5))
sns.regplot(x='release_year', y='year_added', data=df_corr, scatter_kws={'alpha': 0.3}, line_kws={'color': 'red'})
plt.title('Release Year vs Year Added to Netflix')
plt.xlabel('Release Year')
plt.ylabel('Year Added')
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Monthly Content Additions by Type
Understand the release seasonality by type of content.

```python
import matplotlib.pyplot as plt

df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df['month_added'] = df['date_added'].dt.month

monthly_type = df.groupby(['month_added', 'type']).size().unstack().fillna(0)
monthly_type.plot(kind='bar', stacked=True, figsize=(12,6), colormap='Dark2')
plt.title('Monthly Additions by Content Type')
plt.xlabel('Month')
plt.ylabel('Number of Titles')
plt.xticks(ticks=range(12), labels=['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'], rotation=45)
plt.legend(title='Type')
plt.tight_layout()
plt.show()
```

# Questions

```python
# 1ï¸âƒ£ What is the most common genre (listed first in the 'listed_in' column)?
df['main_genre'] = df['listed_in'].str.split(',').str[0]
top_genres = df['main_genre'].value_counts().head(10)
plt.figure(figsize=(10,5))
sns.barplot(x=top_genres.values, y=top_genres.index, palette='viridis')
plt.title('Top 10 Most Common Genres')
plt.xlabel('Number of Titles')
plt.ylabel('Main Genre')
plt.show()
print("Most common genre:", top_genres.idxmax())
```

```python
# 2ï¸âƒ£ Which year had the highest number of releases?
top_release_years = df['release_year'].value_counts().sort_index(ascending=True)
plt.figure(figsize=(12,6))
top_release_years.plot(kind='line', color='green')
plt.title('Number of Titles Released Each Year')
plt.xlabel('Release Year')
plt.ylabel('Number of Titles')
plt.grid(True)
plt.show()
print("Year with most releases:", top_release_years.idxmax())
```

```python
# 3ï¸âƒ£ What are the top 10 most frequent cast members?
cast_exploded = df[df['cast'] != 'Not Specified']['cast'].str.split(',').explode().str.strip()
top_cast = cast_exploded.value_counts().head(10)
plt.figure(figsize=(10,5))
sns.barplot(x=top_cast.values, y=top_cast.index, palette='coolwarm')
plt.title('Top 10 Most Frequent Cast Members')
plt.xlabel('Number of Appearances')
plt.ylabel('Cast Member')
plt.show()
print("Most frequent cast member:", top_cast.idxmax())
```

```python
# 4ï¸âƒ£ Which countries produce the most movies vs. TV shows?
type_country = df.groupby(['country', 'type']).size().unstack().fillna(0)
top_type_country = type_country.sort_values(by='Movie', ascending=False).head(10)
top_type_country[['Movie', 'TV Show']].plot(kind='bar', stacked=True, figsize=(12,6), colormap='Accent')
plt.title('Top 10 Countries by Type of Content')
plt.xlabel('Country')
plt.ylabel('Number of Titles')
plt.legend(title='Type')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 5ï¸âƒ£ What months see the most content added to Netflix?
monthly_additions = df['month_added'].value_counts().sort_index()
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
          'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
plt.figure(figsize=(10,5))
sns.barplot(x=months, y=monthly_additions.values, palette='cubehelix')
plt.title('Content Additions by Month')
plt.xlabel('Month')
plt.ylabel('Number of Titles Added')
plt.show()
```
